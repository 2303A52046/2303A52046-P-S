```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("/content/cricket_dataset.csv")
data.head(50)
```

```python
def plot_histogram(data, column, bins=10):
    plt.figure(figsize=(10, 6))
    sns.histplot(data[column], bins=bins, kde=True, color='blue')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
def plot_scatter_plot(data, x_column, y_column):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=data[x_column], y=data[y_column], color='red')
    plt.title(f'Scatter Plot of {x_column} vs {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()
```

```python
def plot_bar_chart(data, x_column, y_column):
    plt.figure(figsize=(10, 6))
    sns.barplot(x=data[x_column], y=data[y_column], color='green')
    plt.title(f'Bar Chart of {x_column} by {y_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(rotation=45)
    plt.show()
```

```python
file_path = "/content/cricket_dataset.csv"
data = pd.read_csv(file_path)
```

```python
plot_histogram(data, 'Runs')
plot_histogram(data, 'Wickets')
```

```python
plot_scatter_plot(data, 'Matches', 'Runs')
plot_scatter_plot(data, 'Matches', 'Wickets')
```

```python
plot_bar_chart(data, 'Player', '50s')
plot_bar_chart(data, 'Player', '100s')
```
