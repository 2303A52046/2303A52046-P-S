```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/student_data_250.csv")
# Histogram of GPA
plt.figure(figsize=(8, 5))
sns.histplot(df['GPA'], bins=10, kde=True, color='blue')
plt.xlabel("GPA")
plt.ylabel("Number of Students")
plt.title("Distribution of GPA")
plt.show()
```

```python
# Scatterplot of GPA vs. Attendance Rate
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Attendance Rate (%)'], y=df['GPA'], alpha=0.7)
sns.regplot(x=df['Attendance Rate (%)'], y=df['GPA'], scatter=False, color='red')
plt.xlabel("Attendance Rate (%)")
plt.ylabel("GPA")
plt.title("GPA vs. Attendance Rate")
plt.show()
```

```python
# Bar chart of student count per Major
plt.figure(figsize=(10, 5))
df['Major'].value_counts().plot(kind='bar', color='green')
plt.xlabel("Major")
plt.ylabel("Number of Students")
plt.title("Number of Students per Major")
plt.xticks(rotation=45)
plt.show()
```

```python
# Grouped Bar Chart: Extracurricular Activities vs Graduation Year
plt.figure(figsize=(10, 5))
sns.countplot(x='Graduation Year', hue='Extracurricular Activities', data=df)
plt.xlabel("Graduation Year")
plt.ylabel("Number of Students")
plt.title("Extracurricular Activities by Graduation Year")
plt.legend(title="Extracurricular Activities")
plt.show()
```

```python
# Analysis
print("Most common GPA range:", df['GPA'].value_counts(bins=5).idxmax())
print("Correlation between attendance and GPA:", df[['Attendance Rate (%)', 'GPA']].corr().iloc[0, 1])
print("Major with highest enrollment:", df['Major'].value_counts().idxmax())
print("Average attendance rate by extracurricular involvement:")
print(df.groupby('Extracurricular Activities')['Attendance Rate (%)'].mean())
```
