Normal Population and Shape of Distributions

Question 1:
Generate a normal distribution of data points with a mean of 60 and a standard
deviation of 12. Plot a histogram for the generated data. Analyze the symmetry and
bell-shaped curve of a normal distribution and visualize its Probability Density
Function (PDF).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


mean = 60
std_dev = 12


data = np.random.normal(mean, std_dev, 1000)


plt.figure(figsize=(10, 6))
sns.histplot(data, bins=30, kde=True, color='blue')
plt.title('Histogram of Generated Data with KDE')
plt.xlabel('Data Points')
plt.ylabel('Frequency')


plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.legend()
plt.show()


plt.figure(figsize=(10, 6))
sns.kdeplot(data, color='blue', fill=True)
plt.title('Probability Density Function (PDF) of Generated Data')
plt.xlabel('Data Points')
plt.ylabel('Density')
plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.legend()
plt.show()
```

Question 2:
The normalized distribution for a person’s daily calorie intake has a mean of 30
calories and a standard deviation of 5 calories. Write a python code that Calculates and
visualizes the following probabilities for a randomly selected person having a daily
calorie intake:
I. Greater than or equal to 19 calories.
II. Less than or equal to 19 calories.
III. Between 19 and 35 calories.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns


mean = 30
std_dev = 5


x_values = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)


pdf_values = stats.norm.pdf(x_values, mean, std_dev)


prob_greater_equal_19 = 1 - stats.norm.cdf(19, mean, std_dev)
prob_less_equal_19 = stats.norm.cdf(19, mean, std_dev)
prob_between_19_and_35 = stats.norm.cdf(35, mean, std_dev) - stats.norm.cdf(19, mean, std_dev)


print(f"Probability of daily calorie intake >= 19 calories: {prob_greater_equal_19:.4f}")
print(f"Probability of daily calorie intake <= 19 calories: {prob_less_equal_19:.4f}")
print(f"Probability of daily calorie intake between 19 and 35 calories: {prob_between_19_and_35:.4f}")


plt.figure(figsize=(10, 6))
sns.lineplot(x=x_values, y=pdf_values, color='blue')


x_fill = np.linspace(19, mean + 4*std_dev, 1000)
y_fill = stats.norm.pdf(x_fill, mean, std_dev)
plt.fill_between(x_fill, y_fill, color='green', alpha=0.5, label=f'P(X >= 19) = {prob_greater_equal_19:.4f}')


x_fill = np.linspace(mean - 4*std_dev, 19, 1000)
y_fill = stats.norm.pdf(x_fill, mean, std_dev)
plt.fill_between(x_fill, y_fill, color='orange', alpha=0.5, label=f'P(X <= 19) = {prob_less_equal_19:.4f}')


x_fill = np.linspace(19, 35, 1000)
y_fill = stats.norm.pdf(x_fill, mean, std_dev)
plt.fill_between(x_fill, y_fill, color='purple', alpha=0.5, label=f'P(19 <= X <= 35) = {prob_between_19_and_35:.4f}')

plt.title('Probability Density Function (PDF) of Daily Calorie Intake')
plt.xlabel('Daily Calorie Intake')
plt.ylabel('Density')
plt.axvline(x=19, color='red', linestyle='dashed', linewidth=2, label='Calorie Intake = 19')
plt.axvline(x=35, color='red', linestyle='dashed', linewidth=2, label='Calorie Intake = 35')
plt.legend()
plt.show()
```
